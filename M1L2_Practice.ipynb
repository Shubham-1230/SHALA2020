```
n = np.arange(-100,100)
mean = 0
normal = stats.norm.pdf(n,mean,25)
plt.plot(n,normal)
```

```
n = np.arange(-100,100)
mean = 0
normal = stats.norm.cdf(n,mean,25)
plt.plot(n,normal)
```

Poission distribution

```
n = np.arange(0,50)
for Lambda in range(0,10,2):
    poission = stats.poisson.pmf(n,Lambda)
    plt.plot(n,poission,'-o',label='λ = {:f}'.format(Lambda))
    plt.legend()
```

Exponential distribution

```
Lambda = 0.5
x = np.arange(0,15,0.1)
y = Lambda*np.exp(-Lambda*x)
plt.plot(x,y,label='λ={:f}'.format(Lambda))
plt.legend()
```
